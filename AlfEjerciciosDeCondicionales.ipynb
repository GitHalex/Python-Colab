```python
renta_anual = int(input('Ingrese su renta anual: '))
if renta_anual < 10000:
  print('Tipo impositivo es 5%')
elif 10000 >= renta_anual <= 20000:
  print('Tipo impositivo es 15%')
elif 20000 >= renta_anual <= 35000:
  print('Tipo impositivo 20%')
elif 35000 >= renta_anual <= 60000:
  print('Tipo impositivo es 30% ')
else:
  print('Tipo impositivo es 45%')
```

```python
puntuacion = float(input('''
Ingrese su puntuacion

Inaceptable: 0.0
Aceptable: 0.4
Meritorio: 0.6 o mas

'''))
if puntuacion == 0.0:
  print(f'Cantidad conseguida: {0.0*2400}$ Nivel Inaceptable')
elif puntuacion == 0.4:
  print(f'Cantidad conseguida: {0.4*2400}$ Nivel Aceptable')
elif puntuacion > 0.6:
  print(f'Cantidad conseguida: {0.6*2400}$ Meritorio')
else:
  print('No ingresaste una puntuacion correcta')
```
